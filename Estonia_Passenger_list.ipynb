## Classifier

As we know, our data is highly unbalanced and therefore we need to use some techniques to prevent our classifier from being biased.

Our data set is not large and, therefore, we will not use Undersample techniques, instead we will use an Oversampling technique called SMOTE, but this technique makes overfitting more likely.

Finally in this notebook I used a technique called cross-validation to run our modeling process on different subsets of the data to get multiple measures of model quality.

```python
#Using StratifiedKFold to get our subsets

X = df.drop('Survived', axis=1)
y = df['Survived']

sss = StratifiedKFold(n_splits=5, random_state=None, shuffle=False)

for train_index, test_index in sss.split(X, y):
    original_Xtrain, original_Xtest = X.iloc[train_index], X.iloc[test_index]
    original_ytrain, original_ytest = y.iloc[train_index], y.iloc[test_index]
    
original_Xtrain = original_Xtrain.values
original_Xtest = original_Xtest.values
original_ytrain = original_ytrain.values
original_ytest = original_ytest.values
```

```python
#Setting the classifiers

# Logistic Regression 
log_reg_params = {"penalty": ['l1', 'l2'], 'C': [0.001, 0.01, 0.1, 1, 10, 100, 1000]}
grid_log_reg = GridSearchCV(LogisticRegression(), log_reg_params)

# KNears 
knears_params = {"n_neighbors": list(range(2,5,1)), 'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute']}
grid_knears = GridSearchCV(KNeighborsClassifier(), knears_params)

# Support Vector Classifier
svc_params = {'C': [0.5, 0.7, 0.9, 1], 'kernel': ['rbf', 'poly', 'sigmoid', 'linear']}
grid_svc = GridSearchCV(SVC(), svc_params)

# DecisionTree Classifier
tree_params = {"criterion": ["gini", "entropy"], "max_depth": list(range(2,4,1)), 
              "min_samples_leaf": list(range(5,7,1))}
grid_tree = GridSearchCV(DecisionTreeClassifier(), tree_params)

#XGBC
params = {
        'min_child_weight': [1, 5, 10],
        'gamma': [0.5, 1, 1.5, 2, 5],
        'subsample': [0.6, 0.8, 1.0],
        'colsample_bytree': [0.6, 0.8, 1.0],
        'max_depth': [3, 4, 5]
        }
grid_xgbc = GridSearchCV(XGBClassifier(), params)

#randomForest

param_grid = { 
    'n_estimators': [200, 500],
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_depth' : [4,5,6,7,8],
    'criterion' :['gini', 'entropy']
}

grid_random = GridSearchCV(RandomForestClassifier(), param_grid)


classifiers = {
    "Logistic Regression": grid_log_reg,
    "KNearest": grid_knears,
    "Support Vector Classifier": grid_svc,
    "DecisionTreeClassifier": grid_tree,
    "XGBC": grid_xgbc,
    "RandomForest": grid_random
}
```

```python
#SMOTE with Cross-validation

accuracy_lst = []
precision_lst = []
recall_lst = []
f1_lst = []
auc_lst = []
labels = ['not Survived', 'Survived']
result_classifier={}

for key, classifier in classifiers.items():
    rand = classifier
    for train, test in sss.split(original_Xtrain, original_ytrain):
        pipeline = imbalanced_make_pipeline(SMOTE(sampling_strategy='minority'), rand) # SMOTE happens during Cross Validation not before..

        model = pipeline.fit(original_Xtrain[train], original_ytrain[train])
        best_est = rand.best_estimator_
        prediction = best_est.predict(original_Xtrain[test])

        accuracy_lst.append(pipeline.score(original_Xtrain[test], original_ytrain[test]))
        precision_lst.append(precision_score(original_ytrain[test], prediction))
        recall_lst.append(recall_score(original_ytrain[test], prediction))
        f1_lst.append(f1_score(original_ytrain[test], prediction))
        auc_lst.append(roc_auc_score(original_ytrain[test], prediction))
    
    
    result_classifier[key] = best_est
    print('---' * 45)
    print(key)
    print("accuracy: {}".format(np.mean(accuracy_lst)))
    print("precision: {}".format(np.mean(precision_lst)))
    print("recall: {}".format(np.mean(recall_lst)))
    print("f1: {}".format(np.mean(f1_lst)))
    print('Test set')
    smote_prediction = best_est.predict(original_Xtest)
    print(classification_report(original_ytest, smote_prediction, target_names=labels))
    print('---' * 45)
    
    accuracy_lst = []
    precision_lst = []
    recall_lst = []
    f1_lst = []
    auc_lst = []
```

```python
#confusion_matrix

fig, ax = plt.subplots(3, 2,figsize=(50,25))
i = 0
j = 0
for key, model in result_classifier.items():

    y_score = model.predict(original_Xtest)
    roc_auc = roc_auc_score(original_ytest, y_score)

    matrix = confusion_matrix(original_ytest, y_score)
    
    sns.heatmap(matrix, ax=ax[i][j], annot=True, cmap=plt.cm.copper)
    sns.set(font_scale=2)
    ax[i, j].set_title( key + " Confusion Matrix "+ ' AUC: {0:0.2f}'.format(roc_auc), fontsize=20)
    ax[i, j].set_xticklabels(["Not Survived","Survived"], fontsize=20, rotation=0)
    ax[i, j].set_yticklabels(["Not Survived","Survived"], fontsize=20, rotation=360)
    ax[i, j].set_ylabel('True label',fontsize=20,)
    ax[i, j].set_xlabel('Predicted label',fontsize=20,)
    
    if i < 2:
        i+=1
    else:
        i = 0
        j = 1
        
plt.show()
```

As we can see, our model is not simply saying that everyone did not survive, as expected from a biased model. But we still need to make some improvements to get a higher score.
